```python
import numpy as np
import pandas as pd
import os
```

```python
# Get raw data
def get_raw_data(path):
    df = pd.read_excel(path)
    #print(df.head(3))
    return df
```

```python
# Load data
raw_data = get_raw_data('../data/raw_data/cascaded_use_case_data.xlsx')
```

```python
# Review the data
raw_data.describe()
```

```python
# What dataset contains
print(raw_data.columns)
```

```python
# Find amount of null values in dataset
raw_data.info()
```

```python
inflow_Kalltveit = raw_data[['Q_Kalltveit']]
inflow_Kalltveit.index = raw_data['Datetime']
inflow_Kalltveit.describe()
```

```python
inflow_Kalltveit_uten_tapping =  raw_data[['Q_Kalltveit_uten_tapping']]
inflow_Kalltveit_uten_tapping.index = raw_data['Datetime']
inflow_Kalltveit_uten_tapping.describe()
```

```python
from sklearn.model_selection import train_test_split

train, temp = train_test_split(inflow_Kalltveit, train_size=0.7, shuffle=False)
val, test = train_test_split(temp, test_size=0.3, shuffle=False)

print(train.shape, val.shape, test.shape)
```

```python
train.to_csv(r'../data/clean_data/univariate_Q_Kallveit/train.csv')
val.to_csv(r'../data/clean_data/univariate_Q_Kallveit/val.csv')
test.to_csv(r'../data/clean_data/univariate_Q_Kallveit/test.csv')
```
